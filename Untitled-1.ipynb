Now I split the data into a test and train set with a 80-20 distribution

```python
import numpy as np

np.random.seed(42)
def shuffle_and_split_data(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
#train_set, test_set = shuffle_and_split_data(df_oneHotEncoded, 0.2) #Uncomment to see results with one hot encoding
train_set, test_set = shuffle_and_split_data(df, 0.2) #Uncomment to see results without one hot encoding
#print(len(train_set))
#print(len(test_set))
y_train = train_set['Target']#Splits the data into a train set and a  test set
x_train = train_set.drop(columns=['Target'])
y_test = test_set['Target']
x_test = test_set.drop(columns=['Target'])
#print(x_train.head)
```

It was shown previously that the data is imbalanced, Therefore I use SMOTE to create synthetic data

```python
from imblearn.over_sampling import SMOTE


smote = SMOTE(sampling_strategy='auto', random_state=42)

# Apply SMOTE to create synthetic samples
x_SMOTED_train, y_SMOTED_train = smote.fit_resample(x_train, y_train)
x_SMOTED_test, y_SMOTED_test = smote.fit_resample(x_test, y_test)

# Display distribution before and after SMOTE
print("Train distribution before SMOTE:", {label: count for label, count in zip(*np.unique(y_train, return_counts=True))})
print("Train distribution after SMOTE:", {label: count for label, count in zip(*np.unique(y_SMOTED_train, return_counts=True))})
print("Test distribution before SMOTE:", {label: count for label, count in zip(*np.unique(y_test, return_counts=True))})
print("Test distribution after SMOTE:", {label: count for label, count in zip(*np.unique(y_SMOTED_test, return_counts=True))})
```

```python
from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

rnd_clf = RandomForestClassifier(n_estimators=500, random_state=42)
iris = load_iris(as_frame=True)
rnd_clf.fit(x_SMOTED_train, y_SMOTED_train)
for score, name in zip(rnd_clf.feature_importances_, x_train.columns):
    print(round(score, 2), name)#calculates importance of each feature
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score

log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(x_SMOTED_train, y_SMOTED_train)
y_pred_log_reg = log_reg.predict(x_SMOTED_test)

accuracy = accuracy_score(y_SMOTED_test, y_pred_log_reg)
f1 = f1_score(y_SMOTED_test, y_pred_log_reg, average=None) #micro macro or weighted 
print("Accuracy on test data:", accuracy)#random forest classification results
print("f1 score =", f1, "(Graduate, Dropout, Enrolled)")
```
